# Candidate Policies Discovery

Based on the results of MS-MORDM, we would select the candidate policies. Two scenarios were used in MS-MORDM as references, Scenario 0 (file: MORDM S0A.csv) and Scenario 1 (file: MORDM S1A.csv). The policies provided in those two files were the results of MORDM with zero Total A.4 Expected Annual Damage and zero Total A.4 Expected Number of Deaths as constraints. Here, two policies would be selected from each scenario:
1. The effective:
    
    The policy which has the minimum Total A.4 Expected Annual Damage, minimum Total A.4 Expected Number of Deaths and minimum Total Costs (evacuation costs, dike heightening costs, and RfR costs). The Total Costs is the cost burden for Gorssel only. So, the RfR costs and evacuation costs here are assumed to be equally shared between all dike rings.
    
    
2. The least dependency:

    The policy which has the minimum Total A.4 Expected Annual Damage, minimum Total A.4 Expected Number of Deaths and minimum dependency to other regions (minimum RfR and dike heightening in other regions)

```python
#FUNCTIONS FOR AGGREGATING AND DISAGGREGATING THE OOIS
def aggregate_outcomesA4(results, outcome):
    '''
    Aggregating the outcomes related to Gorssel (A.4)
    
    Parameters
    ----------
    results : dataframe
              the data from which the outcomes will be aggregated 
              and the result of aggregation will be appended on
    outcome : str
              the word(s) that characterize the outcome
              we want to aggregate, and the word(s) that will 
              be used as the name of the aggregate outcome 

    Return
    ----------
    results : dataframe
              the data from with appended aggregation result in the 
              last column
    '''
    
    a4 = "A.4"
    list_outcomes_columns = []

    for i in results.columns:
        if a4 in i:
            if outcome in i:
                list_outcomes_columns.append(i)

    results["Total A.4 " + str(outcome)
            ] = results[list_outcomes_columns].sum(axis=1)

def aggregate_outcomes(results, outcome):
    '''
    Aggregating the outcomes regardless the location of the dike
    
    Parameters
    ----------
    results : dataframe
              the data from which the outcomes will be aggregated 
              and the result of aggregation will be appended on
    outcome : str
              the word(s) that characterize the outcome
              we want to aggregate, and the word(s) that will 
              be used as the name of the aggregate outcome 
                
    Return
    ----------
    results : dataframe
              the data from with appended aggregation result in the 
              last column
    '''
    
    list_outcomes_columns = []

    for i in results.columns:
          if outcome in i:
                list_outcomes_columns.append(i)

    results["Total " + str(outcome)
            ] = results[list_outcomes_columns].sum(axis=1)
    
def aggregatendisaggregate_outcomes(results, outcome):
    list_outcomes_columns = []
    '''
    Aggregating the outcome regardless the dike location over 
    planning steps, then equally divide with 5 (number of dike rings).
    Particularly used to calculate Room for River or Evacuation costs 
    to be carried out by Gorssel.
    
    Parameters
    ----------
    results : dataframe
              the data from which the outcomes will be aggregated 
              and the result of aggregation will be appended on
    outcome : str
              the word(s) that characterize the outcome
              we want to aggregate, and the word(s) that will 
              be used as the name of the aggregate outcome 
      
    Return
    ----------
    results : dataframe
              the data from with appended aggregation result in the 
              last column
    '''
    
    for i in results.columns:
          if outcome in i:
                list_outcomes_columns.append(i)

    results["Total A.4" + str(outcome)
            ] = results[list_outcomes_columns].sum(axis=1)/5
    
    

def dependencyRfR(results, lever):
    '''
    Calculating the total number of RfR implementation in dike rings other than 
    Gorssel.
    Considering that RfR is far more cumbersome than Dike heightening, 
    it is weighed 4 times than dependency to Dike Heightening.

    Parameters
    ----------
    results : dataframe
              the data from which the outcomes will be aggregated 
              and the result of aggregation will be appended on
    lever   : str
              the word(s) that characterize the outcome
              we want to aggregate, and the word(s) that will 
              be used as the name of the aggregate outcome 
    
    Return
    ----------
    results : dataframe
              the data from with appended aggregation result in the 
              last column
    '''
    
    a4 = "3"
    list_levers_columns = []

    for i in results.columns:
        if a4 not in i:
            if levers in i:
                list_levers_columns.append(i)
                        
    results[ str(levers)+ "in other regions"
            ] = results[list_levers_columns].sum(axis=1)*4 #weighed 4 times higher than DH
    
    
def dependencyDH(results, lever):
    '''
    Calculating the total dike increase in decimeter required in 
    dike rings other than Gorssel.

    Parameters
    ----------
    results : dataframe
              the data from which the outcomes will be aggregated 
              and the result of aggregation will be appended on
    lever   : str
              the word(s) that characterize the outcome
              we want to aggregate, and the word(s) that will 
              be used as the name of the aggregate outcome 
    
    Return
    ----------
    results : dataframe
              the data from with appended aggregation result in the 
              last column
    '''
    
    a4 = "A.4"
    list_levers_columns = []
    
    for i in results.columns:
        if a4 not in i:
            if levers in i:
                list_levers_columns.append(i)
                        
    results[ str(levers)+ "in other regions"
            ] = results[list_levers_columns].sum(axis=1)
```

### Selecting candidate policies from Scenario 0

```python
#loading datasets from MS-MORDM
polS0= pd.read_csv('intermediatedata\\MORDM_S0A.csv', index_col=0)
polS0
```

```python
polS0A4 = polS0
aggregate_outcomesA4(polS0A4, "Expected Annual Damage")
aggregate_outcomesA4(polS0A4, "Expected Number of Deaths")
aggregate_outcomesA4(polS0A4, "Dike Investment Costs")
aggregatendisaggregate_outcomes(polS0A4, "RfR Total Costs")
aggregatendisaggregate_outcomes(polS0A4, "Expected Evacuation Costs")
```

```python
#the total cost for Gorssel
polS0A4['Total A.4 Costs']= (polS0A4['Total A.4 Dike Investment Costs']
                             + polS0A4['Total A.4RfR Total Costs']
                             + polS0A4['Total A.4Expected Evacuation Costs'])
polS0A4
```

```python
dependencyRfR(polS0A4, "_RfR")
dependencyDH(polS0A4, "DikeIncrease")
polS0A4
```

```python
polS0A4['dependency'] = (polS0A4['_RfRin other regions']
                         +polS0A4['DikeIncreasein other regions'])
polS0A4
```

##### Choosing the effective policy from Scenario 0

```python
polS0A4['Total A.4 Expected Annual Damage'].mean()
polS0A4['Total A.4 Expected Number of Deaths'].mean()
```

All policies offered by MORDM of Scenario 0 provide zero Total A.4 Expected Annual Damage and zero Total A.4 Expected Number of Deaths. Therefore, next in selecting the effective policy, we only need to look at the Total Cost.

```python
effectivepol0 = polS0A4.loc[polS0A4['Total A.4 Costs'].idxmin()]
effectivepol0
```

### The selected candidate policies

```python
candidatepol = pd.concat([effectivepol0, effectivepol1, leastdependpol0, leastdependpol1], axis = 1, sort = False)
candidatepol = candidatepol.T
candidatepol
```

```python
candidatepol.to_csv('candidate_policies.csv')
```

The candidate policies promise zero Total A.4 Expected Annual Damage, zero Total A.4 Expected Number of Deaths, Total A.4 Costs from 4.751268e+07	to 2.585600e+08, only 1-3 RfR implementation required in other regions, and 26-56 dm Dike Increase required in other regions. But as seen from the Parcoord above, there might be a trade-off between the Costs and the dependency to other regions.
